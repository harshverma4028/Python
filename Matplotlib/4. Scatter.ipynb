# scatter plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[2,4,5,7,7,9,4]
colors=["r","y","g","b","r"]
plt.scatter(x,y,color="r")


plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day=[1,2,3,4,5,6,7]
no=[2,4,5,7,7,9,4]
no2=[7,3,2,5,2,6,2]
# colors = ["r", "y", "g", "b", "r", "c", "m"]
colors=[10,20,30,40,50,22,32]
sizes=[500,400,200,100,300,300,800]
plt.scatter(day,no,c=colors,cmap="viridis",s=sizes,alpha=0.4,marker="*",edgecolor="g",linewidth=2)
plt.scatter(day,no2,color="r",s=sizes,alpha=1)


plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```
