```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.2,0.0,0.0,0.0]
c=["hotpink","c","b","black"]

plt.pie(x,labels=y,explode=ex,colors=c,startangle=180,shadow=True,autopct="%0.1f%%",radius=1.3,labeldistance=0.8,
       textprops={"fontsize":15, "color": "green"},counterclock=False,wedgeprops={"linewidth":4},rotatelabels=False)

plt.legend(loc=3,title="Programing Language")
plt.show()


#,"edgecolor":"m"
#,"width":2
```
