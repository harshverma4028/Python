# Melt ()

```python
import pandas as pd

var=pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,11,12,13,14,15],"math":[17,18,19,29,21,22]})
var
```

```python
pd.melt(var,id_vars=["days"],var_name="python",value_name="vidrohi")
```

```python
var=pd.DataFrame({"days":[1,1,1,1,2,2],"st_name":["a","b","a","a","b","b"],"eng":[10,11,12,13,14,15],"math":[17,18,19,29,21,22]})
var
```

```python
var.pivot(index="days",columns="st_name",values="eng")
```

```python
var.pivot_table(index="st_name",columns="days",
               aggfunc="sum",margins=True)
```
